```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("county_level_errors.csv")
plt.scatter(x=df['ActualVotes'], y=df['Error'], alpha=0.2, c=df['Correct'])
```

```python
nullFilter = df.Error.notnull()
dfNotNull = df[nullFilter]
print(df.shape, dfNotNull.shape)
plt.hist(x=dfNotNull['Error'], bins=20)
```

```python
df['Error'].describe()
```

```python
df[df.Correct == True]['Error'].describe()
```

```python
df[df.Correct == False]['Error'].describe()
```

```python
df['AbsoluteError'] = abs(df['Error'])
plt.scatter(x=df['ActualVotes'], y=df['AbsoluteError'], alpha=0.2, c=df['Correct'])
```

```python
nullFilter = df.Error.notnull()
dfNotNull = df[nullFilter]
print(df.shape, dfNotNull.shape)
plt.hist(x=dfNotNull['AbsoluteError'], bins=20)
```
